```python
import pandas as pd
import numpy as np
print("Food analysis tools loaded successfully!")

# Creating a simple restaurant dataset
import io
restaurant_data = pd.DataFrame({
    'restaurant_name': ['Nasi Lemak Wanjo', 'Roti Canai Station', 'Hainanese Chicken Rice', 'Mamak Corner', 'Satay Junction'],
    'cuisine_type': ['Malay', 'Indian', 'Chinese', 'Mamak', 'Malay'],
    'price_range': [2, 1, 3, 2, 3],  # 1=low, 2=medium, 3=high
    'customer_rating': [4.5, 4.2, 4.7, 4.0, 4.3],
    'location': ['Kuala Lumpur', 'Penang', 'Johor Bahru', 'Kuala Lumpur', 'Malacca']
})
food_data = restaurant_data
```

```python
# Compiling code from above in single current cell
#First look into the dataset

# See the first 5 rows
print("First 5 restaurants:")
print(food_data.head())
print("\n" + "="*60)

# How big is our data?
print(f"Dataset shape: {food_data.shape}")
print(f"That means {food_data.shape[0]} food items and {food_data.shape[1]} characteristics")
print("\n" + "="*60)

# What informations do we have?
print("Column names and data types:")
print(food_data.info())

# Viewing the cuisine variety
print("Cuisine types available:")
print(food_data['cuisine_type'].value_counts())
print("\n" + "="*60)

# Price analysis
print("Price range analysis:")
print(food_data['price_range'].describe())
print("\n" + "="*60)

# Rating analysis
print("Customer rating statistics:")
print(food_data['customer_rating'].describe())

# Highest rated restaurant
food_data.loc[food_data['customer_rating'].idxmax()]
```

```python
# How do different cuisines compare?
print("\n🔍 COMPARING CUISINE TYPES:")

# Group by cuisine and calculate averages
cuisine_stats = food_data.groupby('cuisine_type').agg({
    'customer_rating': ['mean', 'count'],
    'price_range': 'mean'
}).round(2)

print(cuisine_stats)
print("\n" + "="*60)
```

```python
# The best value cuisine (high rating, reasonable price)
print("\n VALUE ANALYSIS - Rating per Price Unit:")
for cuisine in food_data['cuisine_type'].unique():
    cuisine_subset = food_data[food_data['cuisine_type'] == cuisine]
    if len(cuisine_subset) > 0:
        avg_rating = cuisine_subset['customer_rating'].mean()
        avg_price = cuisine_subset['price_range'].mean()
        value_score = avg_rating / avg_price if avg_price > 0 else 0
        print(f"{cuisine}: Rating {avg_rating:.2f}/Price {avg_price:.1f} = Value Score {value_score:.2f}")

```

```python
# Analysis by location
if 'location' in food_data.columns:
    print("\n LOCATION ANALYSIS:")
    location_stats = food_data.groupby('location').agg({
        'customer_rating': 'mean',
        'price_range': 'mean',
        'restaurant_name': 'count'
    }).round(2)
    location_stats.columns = ['avg_rating', 'avg_price', 'restaurant_count']
    print(location_stats)
```

```python
# Simple visualization
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
locations = food_data['location'].unique()
avg_ratings = [food_data[food_data['location'] == loc]['customer_rating'].mean() for loc in locations]

plt.bar(locations, avg_ratings, color=['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4', '#FFEAA7'])
plt.xlabel('Location')
plt.ylabel('Average Rating')
plt.title('Average Restaurant Ratings by Location')
plt.ylim(3.5, 5.0)
plt.grid(True, alpha=0.3)
plt.show()
```
