```python
# Import SG Airbnb using the link below
# https://data.insideairbnb.com/singapore/sg/singapore/2025-09-28/data/listings.csv.gz

import pandas as pd

df = pd.read_csv("https://data.insideairbnb.com/singapore/sg/singapore/2025-09-28/data/listings.csv.gz")
print("Yes! It's loaded")
```

```python
print("First 5 properties")
print(df.head(5))
```

```python
print("Shape if the data")
print(df.shape)
```

```python
print("Column names")
print(df.columns)
```

```python
# Question 1: What's the top property?
df_sorted = df.sort_values ('review_scores_value', ascending=False)
top_property = df_sorted['name'].iloc[0]
print("Top Property:", top_property)
```

```python
print("Top 5 Properties:")
print(df_sorted['name'].head(5))
```

```python
# Question 2: What's the average score?
property_score = df['review_scores_value'].mean()
print("Property score:", property_score)
```

```python
# Question 3: How many properties have >4.00 satisfaction?
high_statisfaction = df[df['review_scores_value'] > 4.00]
count_high = len(high_statisfaction)
print("Property with > 4.00:", count_high)
```

```python
# More question
# 1. What's the 3rd property in the dataset?
third_property = df_sorted['name'].iloc[2]
print("Third property:", third_property)
```

```python
# 2. What's the highest review score?
max_score = df['review_scores_value'].max
print("Highest review score:", max_score)
```

```python
# 3. BONUS: Can you find any property from a specific neighbourhood?
# Print observation neighbourhood info
print(df[['neighbourhood','neighbourhood_cleansed','neighbourhood_group_cleansed']].head())
```

```python
# 3. BONUS: Can you find any property from a specific neighbourhood?
# Assume we want Central Region of SG
central_region = df['neighbourhood_group_cleansed'] == "Central Region"
col_to_show = ['name','neighbourhood_group_cleansed']
filter_property_name_region = df.loc[central_region, col_to_show]
print("Property Name based on Central Region:")
print(filter_property_name_region.head())
```

```python
# Challenge 1: Create a summary report
print('=== 2025 SINGAPORE AIRBNB SUMMARY ===')
print("Total properties:", len(df))
print("Average rating:", df['review_scores_rating'].mean())
```

```python
# Challenge 2: Find properties with both high rating AND score
excellent = df[(df['review_scores_rating'] > 4.50) & (df['review_scores_value'] > 4.50)]
print("\n Excellent properties (both > 4.50):")
print(excellent[['name', 'review_scores_rating', 'review_scores_value']])
```

```python
# Challenge 3: Add your own rating column
df['my_rating'] = (df['review_scores_rating'] + df['review_scores_value']) / 2
print("\nMy custom rating added!")
print(df[['name', 'my_rating']].head())
```

```python
# BONUS: DIY DataSet

import pandas as pd

simple_data = {
  'properties': ['Lily Residence', 'Teratak Bonda', 'Sleep n Go', 'J Condominium', 'Tower X'],
  'score'     : [1.00, 2.00, 3.00, 4.00, 5.00],
  'rating'    : [1.00, 2.00, 3.00, 4.00, 5.00],
  'city'      : ['JB', 'BP', 'KB', 'Seremban', 'Kuantan']
}

df = pd.DataFrame(simple_data)
print(df)
```

```python
# Question 1: What's the top property?
df_sorted = df.sort_values ('review_scores_value', ascending=False)
top_property = df_sorted['name'].iloc[0]
print("Top Property:", top_property)

# Question 2: What's the average score?
property_score = df['review_scores_value'].mean()
print("Property score:", property_score)

# Question 3: How many properties have >4.00 satisfaction?
high_statisfaction = df[df['review_scores_value'] > 4.00]
count_high = len(high_statisfaction)
print("Property with > 4.00:", count_high)

# More question
# 1. What's the 3rd property in the dataset?
third_property = df_sorted['name'].iloc[2]
print("Third property:", third_property)

# 2. What's the highest review score?
max_score = df['review_scores_value'].max
print("Highest review score:", max_score)

# Challenge 1: Create a summary report
print('=== 2025 SINGAPORE AIRBNB SUMMARY ===')
print("Total properties:", len(df))
print("Average rating:", df['review_scores_rating'].mean())

# Challenge 2: Find properties with both high rating AND score
excellent = df[(df['review_scores_rating'] > 4.50) & (df['review_scores_value'] > 4.50)]
print("\n Excellent properties (both > 4.50):")
print(excellent[['name', 'review_scores_rating', 'review_scores_value']])

# Challenge 3: Add your own rating column
df['my_rating'] = (df['review_scores_rating'] + df['review_scores_value']) / 2
print("\nMy custom rating added!")
print(df[['name', 'my_rating']].head())
```
